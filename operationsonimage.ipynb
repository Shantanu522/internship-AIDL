```python
import cv2
import matplotlib.pyplot as plt
import skimage as ski
```

```python
img=cv2.imread('1.jpg')
img.shape
```

```python
plt.imshow(img[:,:,::-1])
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
th,img_bin=cv2.threshold(img_gray,70,255,cv2.THRESH_BINARY)
img_bin.shape
```

```python
plt.imshow(img_bin,cmap='gray')
```

```python
plt.imshow(img[:200,200:400,::-1])
```

```python
x,y,w,h=(50,150,80,140)
plt.imshow(img[y:y+h,x:x+w,::-1])
```

```python
x1,y1,x2,y2=(50,150,270,210)
plt.imshow(img[y1:y2,x1:x2,::-1])
```

```python
img_large=cv2.resize(img,(1500,1500),interpolation=cv2.INTER_CUBIC)
img_large.shape
```

```python
plt.imshow(img_large[:,:,::-1])
```

```python
img_small=cv2.resize(img,(0,0),fx=0.25,fy=0.25,interpolation=cv2.INTER_CUBIC)
img_small.shape
```

```python
img_90=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_90[:,:,::-1])
```

```python
rot_mat=cv2.getRotationMatrix2D(center=(0,0),angle=30,scale=1.0)
img_x=cv2.warpAffine(img,rot_mat,dsize=img.shape[:2])
plt.imshow(img_x[:,:,::-1])
```

```python
img_x=ski.transform.rotate(img,-30)
plt.imshow(img_x[:,:,::-1])
```

```python
plt.subplot(1,3,1)
plt.title('Original')
plt.imshow(img[:,:,::-1])
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Brightened')
plt.imshow(cv2.add(img[:,:,::-1],+20))
plt.axis('off')

plt.subplot(1,3,3)
plt.title('Darkened')
plt.imshow(cv2.add(img[:,:,::-1],-20))
plt.axis('off')
```

```python
img_cntbrt=cv2.convertScaleAbs(img,alpha=1.25,beta=0)
plt.imshow(img_cntbrt[:,:,::-1])
```

```python
img_cntbrt=cv2.convertScaleAbs(img,alpha=1.25,beta=-140)
plt.imshow(img_cntbrt[:,:,::-1])
```

```python
img_cntbrt=cv2.convertScaleAbs(img,alpha=0.25,beta=0)
plt.imshow(img_cntbrt[:,:,::-1])
```

```python
#alpha->contrast
#beta->brightness
```

```python
img_blur=cv2.medianBlur(img,15)
plt.imshow(img_blur[:,:,::-1])
```

```python
import numpy as np
plt.subplot(1,2,1)
plt.imshow(img[:,:,::-1])
plt.subplot(1,2,2)
sharp_kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
img_sharp=cv2.filter2D(img,ddepth=5,kernel=sharp_kernel)
plt.imshow(img_sharp[:,:,::-1])
```
